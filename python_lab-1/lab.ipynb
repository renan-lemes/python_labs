```python
daily_return = np.log(df4).diff().dropna()
```

```python
daily_return.head(5)
```

```python
daily_return[['AAPL', 'GOOG']].iplot(title='Log do retorno diario ')
```

```python
(daily_return.mean()*252).iplot(kind='bar',title='Retorno anual das ações')
```

```python
rolling_return = daily_return.rolling(5).sum().dropna()
rolling_return.head(5)
```

```python
rolling_return['NFLX'].iplot(title='5 retorno acumulado de 5 dias da ação NetFlix')
```

```python
years = vix.index.year.unique()
newdf = pd.DataFrame()
for year in years:
    newdf[year] = pd.Series(vix[vix.index.year==year]['VIX Close']).reset_index(drop=True)
newdf = newdf.ffill(axis=1)
newdf.head()
```

```python
newdf.describe().T
```
